
## Translating between different language embeddings to make a dictionary

```python
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
import pandas as pd
from sklearn.model_selection import train_test_split
```

## load the model and the data(the data is available on kaggle)


```python
reg = linear_model.LinearRegression()
```

```python
df_infersent = pd.read_csv("/kaggle/input/infersent-cyoyo-75k/infersent_embeddings.csv",index_col=0)
df_use = pd.read_csv("/kaggle/input/use-cyoyo-75k/use_embeddings.csv",index_col=0)
```

```python
prompt = df_infersent['prompt']
df_infersent.drop(['prompt'],axis=1,inplace=True)
df_use.drop(['prompt'],axis=1,inplace=True)
```

```python
X_train,X_val,y_train,y_val = train_test_split(df_infersent,df_use,test_size=0.1)
```

## train the model

```python
reg.fit(X_train,y_train)
```

```python
train_error = mean_squared_error(reg.predict(X_train),y_train)
test_error = mean_squared_error(reg.predict(X_val),y_val)
```

## validation

```python
print(f"train error:{train_error} \ntest_error:{test_error}")
```
