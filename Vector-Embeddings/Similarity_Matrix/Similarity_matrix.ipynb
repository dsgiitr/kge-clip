```python
import pandas as pd 
import numpy as np 
```

```python
clip_image_df= pd.read_csv('/Users/simar/Desktop/Embeddings_csv/Image/Clip_image/clip_embeddings_512.csv')
clip_image_df.drop(columns=['Unnamed: 0'],inplace=True)
clip_image_df
```

```python
clip_text_df=pd.read_csv('/Users/simar/Desktop/Embeddings_csv/Text/Clip_text/clip_text_embeddings.csv')
clip_text_df.drop(columns=['Unnamed: 0'],inplace=True)
clip_text_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image, UnidentifiedImageError
from io import BytesIO
import requests


# Extract the embeddings and metadata
image_embeddings = clip_image_df.iloc[:, :-2].values  # all columns except last two (url, text)
text_embeddings = clip_text_df.iloc[:, :-1].values  # all columns except last one (prompt)
image_urls = clip_image_df['url'].values
text_prompts = clip_text_df['prompt'].values

# Select the first 10 embeddings
image_embeddings_10 = image_embeddings[10:20]
text_embeddings_10 = text_embeddings[10:20]
image_urls_10 = image_urls[10:20]
text_prompts_10 = text_prompts[10:20]

# Calculate cosine similarities for the first 10 embeddings
cosine_sim_matrix = cosine_similarity(image_embeddings_10, text_embeddings_10)

# Function to fetch and display images
def fetch_image(url):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        return img
    except (requests.exceptions.RequestException, UnidentifiedImageError):
        print(f"Could not fetch image from URL: {url}")
        return None

# Fetch images
images = [fetch_image(url) for url in image_urls_10]

# Plotting the heatmap
plt.figure(figsize=(20, 14))
plt.imshow(cosine_sim_matrix, vmin=0.1, vmax=0.3, cmap='viridis')

# Display text prompts on the y-axis
plt.yticks(range(len(text_prompts_10)), text_prompts_10, fontsize=18)

# Display images on the x-axis
plt.xticks([])
for i, image in enumerate(images):
    if image is not None:
        image = image.resize((50, 50), Image.LANCZOS)
        plt.imshow(image, extent=(i - 0.5, i + 0.5, -1.6, -0.6), origin="lower")

# Annotate the heatmap with similarity scores
for x in range(cosine_sim_matrix.shape[1]):
    for y in range(cosine_sim_matrix.shape[0]):
        plt.text(x, y, f"{cosine_sim_matrix[y, x]:.2f}", ha="center", va="center", size=12)

# Customize plot appearance
for side in ["left", "top", "right", "bottom"]:
    plt.gca().spines[side].set_visible(False)

plt.xlim([-0.5, len(text_prompts_10) - 0.5])
plt.ylim([len(text_prompts_10) + 0.5, -2])
plt.title("Cosine Similarity between Text and Image Features", size=20)
plt.show()
```
