```python
import numpy as np
import tensorflow_hub as hub
import pandas as pd
import tensorflow as tf
from sklearn.decomposition import PCA
```

## load model and data

```python
from datasets import load_dataset

dataset = load_dataset("recastai/coyo-75k-augmented-captions")
sentences = dataset['train']['short_caption']
```

```python
sentences = [sentence[0] for sentence in sentences]
```

```python
embed = hub.load("/kaggle/input/universal-sentence-encoder/tensorflow2/universal-sentence-encoder/2")
```

## Code to make the emdeddings 

```python
embeddings = embed(sentences)
```

## Saving the embeddings

```python
df = pd.DataFrame(embeddings)
df['prompt'] = sentences
df.to_csv("use_embeddings.csv")
```
